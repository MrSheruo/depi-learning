```python
import pandas as pd

df = pd.read_csv("./emp.csv")
df 
```

```python
"""
- Load the employees.csv file into a DataFrame.
- Display:
    The first 5 rows of the dataset
    Dataset information (info())
    Statistical summary of numerical columns (describe())
- Check if there are any missing values in the dataset
- Handle missing values using one of the following methods:
    Fill missing salaries with the average salary
    OR
    Drop rows that contain missing values
- Convert the Joining_Date column to a datetime format.
- Display only the Name, Department, and Salary columns.
- Filter and display employees who:
    Work in the IT department
    Are older than 30 years
-Add a new column called Salary_After_Raise that increases each employee’s salary by 5%.
- Sort employees by Age in descending order.
- Calculate The average salary per department
- Calculate The number of employees in each department
- Find all employees who joined after January 1, 2020.
- final save the new data into csv and excel file
 
* Advance
- Identify the top 3 highest-paid employees
- Create the following new columns:
    Experience_Years → number of years since joining (based on current year)
    Age_Group:
        Under 25
        25–35
        36–50
        Above 50
- Identify employees whose salary is:
    Above their department’s average
    Below their department’s average
 
- Create a column called High_Earner:
    Yes if salary > 75th percentile
    No otherwise
 
- Create a column called Senior_Employee:
    Yes if Experience_Years ≥ 5
    No otherwise
"""
```

```python
# - Display:
#     The first 5 rows of the dataset
#     Dataset information (info())
#     Statistical summary of numerical columns (describe())
print(df.head(5))
print(df.info())
print(df.describe())
```

```python
# Check if there are any missing values in the dataset
df.isnull().sum()
```

```python
# - Handle missing values using one of the following methods:
#     Fill missing salaries with the average salary
#     OR
#     Drop rows that contain missing values
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
```

```python
# - Convert the Joining_Date column to a datetime format.
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"])
```

```python
df
```

```python
#  Display only the Name, Department, and Salary columns.
df.loc[:,["Name","Department","Salary"]]
```

```python
# - Filter and display employees who:
#     Work in the IT department
#     Are older than 30 years

df[(df["Department"] == "IT") & (df["Age"] > 30)]
```

```python
# Add a new column called Salary_After_Raise that increases each employee’s salary by 5%.

df["Salary_After_Raise"] = (df["Salary"] * 0.05).astype(int)
df
```

```python
# - Sort employees by Age in descending order.
df.sort_values(by="Age",ascending=False)
# - Calculate The average salary per department
df.groupby("Department")["Salary"].mean()
# - Calculate The number of employees in each department
df.groupby("Department")["ID"].count()
# - Find all employees who joined after January 1, 2020.
df[df["Joining_Date"] > "2020 , 1 , 1"]
# - final save the new data into csv and excel file
```

```python
df.to_csv("./final.csv")
```
