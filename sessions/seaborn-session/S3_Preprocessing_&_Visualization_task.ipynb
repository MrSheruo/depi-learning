**Line Plot to Show Trends Over Time (flights)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'flights' dataset
data = sns.load_dataset('flights')
print(data)
sns.lineplot(data=data, x='year', y='passengers')
plt.title("Passengers over years")
plt.show()
```

**Scatter Plot to Visualize Relationships Between Two Variables (tips)**

```python
# scatter plot
tips = sns.load_dataset("tips")
print(tips)
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")
plt.title("relation between total bill and tip")
plt.show()
```

**Bar Plot to Show Categorical Data  (tips)**

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("total bill by day")
plt.show()



```

**Histogram to Show Data Distribution (tips)**

```python
sns.histplot(x="size", data=tips)
plt.title("distribution of tables size")
plt.show()
```

**Box Plot to Visualize Distribution and Outliers (tips)**

```python
sns.boxplot(x="sex", y="tip", data=tips)
plt.title("Tip distribution by gender")
plt.show()
```

**Violin Plot to Show Density and Distribution (tips)**

```python
sns.violinplot(x="day", y="tip", data=tips,hue="sex")
plt.title("total bill by day")
plt.show()
```

**Pair Plot to Show Pairwise Relationships (tips)**

```python
sns.pairplot(tips,hue="time")
plt.show()
```

**Count Plot to Show the Frequency of Categorical Values (tips)**

```python
sns.countplot(x="day", data=tips)
plt.title("number of orders per day")
plt.show()
```

Joint Plot to Show Distribution and Relationship (tips)

```python
sns.jointplot(x="size", y="tip", data=tips)
plt.show()
```
